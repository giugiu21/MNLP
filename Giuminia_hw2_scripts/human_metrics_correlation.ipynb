```python
import requests
import pandas as pd
#6th dataset finetuning_corrected
dataset_corrected_g_id = "1FYnqIGNfWPcwlMzNNiEDTQlwLeHf9suA"
url_p6 = f"https://drive.google.com/uc?id={dataset_corrected_g_id}"


response_p6 = requests.get(url_p6)
with open('corrected_g.jsonl', 'wb') as f:
    f.write(response_p6.content)


df_corrected_g = pd.read_json('corrected_g.jsonl', lines=True)
```

```python
datasets_gemini = [df_finetuning_gemini, df_llama_gemini, df_llama_2_gemini, df_qwen_gemini, df_qwen_2_gemini]
datasets_p = [df_finetuning_p, df_llama_p, df_llama_2_p, df_qwen_p, df_qwen_2_p]

def highest_score_gemini():
    counts = []
    for dataset in datasets_gemini:
        count_5 = int(dataset['Score'].eq(5.0).sum())
        counts.append(count_5)
    return counts

def highest_score_p():
    counts_p = []
    for dataset in datasets_p:
        count_5 = int(dataset['Score'].eq(5).sum())
        counts_p.append(count_5)
    return counts_p
```

```python

gemini_counts = highest_score_gemini()
print("Count of translations with highest score (Gemini):", gemini_counts)

prometheus_counts = highest_score_p()
print("Count of translations with highest score (Prometheus):", prometheus_counts)

```

```python
def average_score_gemini():
    averages = []
    for dataset in datasets_gemini:
        avg_score = float(dataset['Score'].mean())
        averages.append(avg_score)
    return averages

def average_score_p():
    averages_p = []
    for dataset in datasets_p:
        avg_score = float(dataset['Score'].mean())
        averages_p.append(avg_score)
    return averages_p
```

```python
gemini_averages = average_score_gemini()
print("Average score (Gemini):", gemini_averages)

prometheus_averages = average_score_p()
print("Average score (Prometheus):", prometheus_averages)
```
