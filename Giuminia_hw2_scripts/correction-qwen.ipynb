```python
import pandas as pd
import requests

dataset_finetuning_id = "1shBzevdsxKxGgfRprcK1TwqO3vZdTrTX"
url = f"https://drive.google.com/uc?id={dataset_finetuning_id}"

response = requests.get(url)
with open('finetuning.jsonl', 'wb') as f:
    f.write(response.content)


df_finetuning = pd.read_json('finetuning.jsonl', lines=True)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-14B")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen3-14B", device_map="auto", torch_dtype=torch.float16)
```

```python
from transformers import pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
import re
def translate(sentence, translation):
    prompt = (
        f'Revise the translation of the following sentence in old italian mantaining as much as possible relevance to the original sentence\'s meaning.\n'
        f'Paraphrase the translation using Modern Italian terms, an example of acceptable correction is:'
        f'Old Italian: "Marco Cornelio ch\'era de\' dieci compagni, studiosamente  si riservò di parlare all\'ultimo."'
        f'Translation: "Marco Cornelio, che era dei dieci compagni, si riservò di parlare all\'ultimo con studio."'
        f'Corrected Translation: "Marco Cornelio, che era tra i dieci compagni, scelse di parlare per ultimo intenzionalmente"'
        f'Only return the corrected translation enclosed in double quotes. No notes, no explanations.\n\n'
        f'Old Italian: "{sentence}"\n'
        f'Modern Italian:"{translation}"'
        f'Corrected Translation:'
    )

    output = pipe(prompt, max_new_tokens=100, do_sample=False)[0]["generated_text"]

    # Extract only what follows "Modern Italian:"
    response = output.split("Corrected Translation:")[-1].strip()

    # Extract only what is inside the quotation marks ""
    match = re.search(r'"([^"]+)"', response)

    if match:
        correction = match.group(1).strip()
    else:
        # Fallback: use entire the translation given
        correction = translation

    return correction
```

```python
df_finetuning['Correction'] = df_finetuning.apply(
    lambda row: translate(row['Sentence'], row['Translation']), axis=1
)
```
